```python
url = 'https://api.exchangerate.host/latest'
response = requests.get(url,params = {'base':'USD'})
exchange_rate = response.json()['rates']
exchange_rate
```

```python
rates = pd.DataFrame.from_dict(exchange_rate, orient ='index', columns = ['Exchange_rate'])
rates.index.name = 'Currency code'
rates
```

```python
list(data_modified.columns)
```

```python
real_currency = data_modified[['Annual salary','Additional monetary compensation','The currency']]
currency= real_currency.merge(rates ,how = 'left', left_on = 'The currency', right_index = True)
```

```python
currency['annual salary in USD'] = currency['Annual salary']/currency['Exchange_rate']
currency['annual salary in USD'] = currency['annual salary in USD'].astype(int)
```

```python
currency['Additional monetary compensation in USD'] = currency['Additional monetary compensation']/currency['Exchange_rate']
currency['Additional monetary compensation in USD'] = currency['Additional monetary compensation in USD'].astype(int)
salary_in_usd = currency[['The currency','annual salary in USD','Additional monetary compensation in USD']]
```
