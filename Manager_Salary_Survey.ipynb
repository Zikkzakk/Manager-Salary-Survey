```python
#imports
import pandas as pd
import requests
```

```python
#Reading the survey dataset into the notebook
dataset = pd.read_excel('Ask_A_Manager_Salary_Survey 2021_(Responses)_edited.xlsx', sheet_name = 'Edited-data')
dataset.head()
```

```python
#Getting the copy of dataset
data_modified = dataset.copy()
```

# Fill in Na values
Currenctly the dataset has a lot of Na value and makes analyse difficult to perform. Since the columns that have Na values have different types and characteristics, Na values will be filled with new values that will match the type and characteristics of each column.

```python
#Checking which column has Na value
data_modified.isna().sum()
```

```python
data_modified.drop(columns = 'Race', inplace = True)
data_modified.dropna(subset = ['Industry'], inplace = True)
data_modified['Job title additional context'].fillna(value = 'no additional context', inplace= True)
data_modified['Additional monetary compensation'].fillna(0, inplace = True)
data_modified['Income additional context'].fillna(value = 'no additional context', inplace= True)
data_modified['State in the U.S.'].fillna(value = 'outside U.S.' , inplace = True)
data_modified.fillna(value = 'no information' , inplace = True)
```

```python
data_modified.isna().sum()
```

# Replacing incorrect and inconsistent currency code

```python
pd.set_option('display.max_rows', None)
data_modified['The currency'].value_counts()
```

```python
data_modified['The currency'].replace(to_replace = ['AUD Australian ','Australian Dollars ','AUD/NZD'], value = 'AUD',inplace = True)
data_modified['The currency'].replace(['American Dollars','US Dollar'],'USD',inplace = True)
data_modified['The currency'].replace(['Philippine Pesos','PhP (Philippine Peso)','Philippine Peso','Php','Philippine peso (PHP)'],'PHP',inplace = True)
data_modified['The currency'].replace(['INR (Indian Rupee)','Indian rupees'],'INR',inplace = True)
data_modified['The currency'].replace(['Argentinian peso (ARS)','Argentine Peso','Peso Argentino'],'ARS',inplace = True)
data_modified['The currency'].replace(['Euro'],'EUR',inplace = True)
data_modified['The currency'].replace(['Israeli Shekels','Ils','ILS/NIS','NIS (new Israeli shekel)','ILS (Shekel)'],'ILS',inplace = True)
data_modified['The currency'].replace(['Dkk','dkk','Danish Kroner','DKK '],'DKK',inplace = True)
data_modified['The currency'].replace(['Mexican Pesos','Mexican pesos'],'MXN',inplace = True)
data_modified['The currency'].replace(['China RMB','RMB (chinese yuan)','RMB'],'CNY',inplace = True)
data_modified['The currency'].replace(['THAI  BAHT','Thai Baht '],'THB',inplace = True)
data_modified['The currency'].replace(['PLN (Zwoty)','PLN (Polish zloty)','Polish Złoty'],'PLN',inplace = True)
data_modified['The currency'].replace(['Korean Won ','KRW (Korean Won)'],'KRW',inplace = True)
data_modified['The currency'].replace(['Taiwanese dollars','NTD'],'TWD',inplace = True)
data_modified['The currency'].replace(['Norwegian kroner (NOK)','Nok'],'NOK',inplace = True)
data_modified['The currency'].replace(['croatian kuna'],'HRK',inplace = True)
data_modified['The currency'].replace(['IDR '],'IDR',inplace = True)
data_modified['The currency'].replace(['Czk','czech crowns'],'CZK',inplace = True)
data_modified['The currency'].replace(['Sgd','SGD '],'SGD',inplace = True)
data_modified['The currency'].replace(['BR$','BRL (R$)'],'BRL',inplace = True)
data_modified['The currency'].replace(['Bdt'],'BDT',inplace = True)
data_modified['The currency'].replace(['RM'],'MYR',inplace = True)
data_modified['The currency'].replace([0, 'Equity'], 'to_drop', inplace =True)
```

```python
#Importing exchangerate API using requests
#exchange rate based on USD currency
url = 'https://api.exchangerate.host/latest'
response = requests.get(url,params = {'base':'USD'})
exchange_rate = response.json()['rates']
rates = pd.DataFrame.from_dict(exchange_rate, orient ='index', columns = ['Exchange_rate'])
rates.index.name = 'Currency code'
rates.head()
```

```python
list(data_modified.columns)
```

```python
real_currency = data_modified[['Annual salary','Additional monetary compensation','The currency']]
currency= real_currency.merge(rates ,how = 'left', left_on = 'The currency', right_index = True)
```

```python
currency['annual salary in USD'] = currency['Annual salary']/currency['Exchange_rate']
currency['annual salary in USD'] = currency['annual salary in USD'].astype(int)
```

```python
currency['Additional monetary compensation in USD'] = currency['Additional monetary compensation']/currency['Exchange_rate']
currency['Additional monetary compensation in USD'] = currency['Additional monetary compensation in USD'].astype(int)
salary_in_usd = currency[['The currency','annual salary in USD','Additional monetary compensation in USD']]
```
